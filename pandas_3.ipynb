### Aggregate by multiple groups/functions

```python
df = pd.read_excel("titanic_save.xlsx")
df
```

```python
df.who.unique()
```

```python
df.groupby('who').count()
```

```python
df.groupby('sex').count()
```

```python
df.groupby('sex').sum()
```

```python
len(df.groupby('sex'))
```

```python
df.groupby(['who','sex','pclass']).count()
```

```python
# selecting columns from a data frame using pandas

df[['age','fare']].mean()
```

```python
# summary of a dataset in python 
df.describe()

# customized output of describe function using loc
df.describe().loc[['min','50%','max']]
```

```python
# customized output of describe function using loc
df.describe().loc['min':'max'] 
```
